# Entendo busca semântica

#### Para o caso, vamos persistir os dados 

```python
import chromadb
client = chromadb.PersistentClient(path="./semantic_db")
```

## Collection

```python
collection = client.get_or_create_collection(name="the_collection")
collection.get()
```

### Atributos das collections

#### list_collections

```python
client.list_collections()
```

## Adicionando documentos à collection

```python
collection.add(
    documents=["Python é uma linguagem de programação orientada a objeto.",
               "Rust é uma linguagem de programação funcional.",
               "Postgres é um banco de dados relacional."],
    embeddings=None,
    metadatas=[
        {"principal_caracteristica": "Facilidade"},
        {"principal_caracteristica": "Rapidez"},
        {"principal_caracteristica": "Relacional"}],
    ids=["id1", "id2", "id3"]
)
```

## Realizando perguntas à collection

```python
query_oop = collection.query(
    query_texts=["Qual a linguagem de programação é orientada a objeto?"],
    n_results=2
)

print(query_oop)
```

```python
query_oop
```

```python
query_func = collection.query(
    query_texts=["Qual a linguagem de programação é funcional?"],
    n_results=2
)

print(query_func)
```

```python
query_db = collection.query(
    query_texts=["Qual é um database?"],
    n_results=1
)

print(query_db)
```
