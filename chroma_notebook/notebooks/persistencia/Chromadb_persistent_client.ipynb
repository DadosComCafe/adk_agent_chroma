## Importando o Chromadb

```python
import chromadb
```

## Criando o cliente

```python
client = chromadb.PersistentClient(path="./meu_db")
```

## Criando uma Collection

```python
collection = client.get_or_create_collection(name="the_collection")
```

## Adicionando algum conteúdo à collection

```python
collection.add(
    ids=["0", "1"],
    documents=[
        "Este é um documento a respeito de Python",
        "Este é um documento a respeito de Java"
    ]
)
```

## Pesquisando na collection

```python
results = collection.query(
    query_texts=["Esta é uma consulta a respeito de tecnologia"],
    n_results=2
)
print(results)
```

### Neste caso, há persistência, um db foi criado. Logo, se encerrar e abrir de novo, será possível realizar a query.

```python
import chromadb
client = chromadb.PersistentClient(path="./meu_db")
collection = client.get_or_create_collection(name="the_collection")
```

```python
results = collection.query(
    query_texts=["Esta é uma consulta a respeito de tecnologia"],
    n_results=2
)
print(results)
```

### Como os dados foram persistidos, na pasta onde este notebook está sendo executado, foi criada uma pasta meu_db, onde dentro há um arquivo .sqlite3 e outro .bin, que são os registros desta persistência de dados
